# Import the data and libraries
________

The first thing I need to do is load in the libraries and dataset. 


```python
# modules we'll use
import pandas as pd
import numpy as np

# for Box-Cox Transformation
from scipy import stats

# for min_max scaling
from sklearn.preprocessing import MinMaxScaler

# plotting modules
import seaborn as sns
import matplotlib.pyplot as plt

```

# Feature Scaling
________

>Feature scaling is the method to limit the range of variables so that they can be compared on common grounds.This means that we're transforming our data so that it fits within a specific scale, like 0-100 or 0-1.
There are algorithms like the Principal Component Analysis(PCA) , k-nearest neighbors(Knn) , Linear / Logistic regression (When regularized) where scaling help us avoid skewness towards high magnitude feature.

```python
# generate 1000 data points randomly drawn from an exponential distribution
data = np.random.exponential(size = 1000)
min_max=MinMaxScaler()
original_data = data.reshape(-1,1)
# Using min_max scaler for transformation
scaled_data=min_max.fit_transform(original_data)
# Plotting the graphs before and after scaling the data to compare
fig, ax=plt.subplots(1,2)
sns.distplot(scaled_data, ax=ax[0])
ax[0].set_title("Original data")
sns.distplot(original_data, ax=ax[1])
ax[1].set_title("Scaled Data")
```

>Range of the data has changed from 0 - 8ish to 0 - 1 but the shape remains the same.

# Normalization
_______
 
Normalization is done to change your observations so that they can be described as a normal distribution. 

>**Normalization** is the process where the features are rescaled so that they’ll have the properties of a standard normal distribution with μ=0 and σ=1, where μ is the mean (average) and σ is the standard deviation from the mean. 

In general, you'll only want to normalize your data if you're going to be using a machine learning or statistics technique that assumes your data is normally distributed. Some examples of these include t-tests, ANOVAs, linear regression, linear discriminant analysis (LDA) and Gaussian naive Bayes. 

I am using Box-Cox Transformation to normalize data. A Box Cox transformation is a way to transform non-normal dependent variables into a normal shape. 

```python
# normalize the exponential data with boxcox
normalized_data = stats.boxcox(original_data)

# Plotting the graphs before and after normalizing the data to compare
fig, ax=plt.subplots(1,2)
sns.distplot(original_data, ax=ax[0])
ax[0].set_title("Original Data")
sns.distplot(normalized_data[0], ax=ax[1])
ax[1].set_title("Normalized data")
```

>Before normalizing it was almost L-shaped and now the *shape* of the data has changed. After normalizing it looks more like a "bell curve".

# Feature Scaling vs. Normalization: What's the difference?
____
>In both cases, we're transforming the values of numeric variables so that the transformed data points have specific helpful properties. The difference is that, in scaling, we're changing the *range* of your data while in normalization we're changing the *shape of the distribution* of our data.
